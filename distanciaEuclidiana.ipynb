## Calculo de distancias Euclidianas N dimensiones


```python
import numpy as np
```

```python
def distanciaEuclidiana(array1, array2):
    array3 = array1 - array2
    elevado = array3 * array3
    suma = np.sum(elevado)
    resultado = np.sqrt(suma)
    return resultado
    
```

```python
print(distanciaEuclidiana(4,8))
```

```python
X = np.array([2,4,6,8])
Y = np.array([1,4,7,10])
print(distanciaEuclidiana(X,Y))
```

## Sklearn programando K-Medias(k-means)
En este ejemplo generaremos un algoritmo de clusterización con K-medias. Considera que el algoritmo dará resultados aleatorios cada vez que lo ejecutes, sin embargo el proceso es el mismo.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
def distanciaEuclidiana(valor1, valor2):
    resta = valor1-valor2
    potencia = np.power(resta,2)
    suma = np.sum(potencia)
    resultado = np.sqrt(suma)
    return resultado
```

```python
x,y = make_blobs(n_samples = 1000, centers = 9, cluster_std=1.5)
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
k_medias = KMeans(n_clusters = 9, max_iter = 300, init = 'random')
k_medias.fit(x)
plt.scatter(x[:,0], x[:,1])
plt.scatter(k_medias.cluster_centers_[:,0], k_medias.cluster_centers_[:,1], s=100, c='yellow')
plt.show()
```
